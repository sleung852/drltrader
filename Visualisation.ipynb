```python
def load_model(model_dir):
    df = pd.read_csv(f'result/{model_dir}/test_result.csv')
    # cumulative reutrn
    df['cum_return'] = ((df['reward']/100+1).cumprod()-1)*100
    df['return'] = df['reward']/100
    annualised_return = (df['reward']/100).add(1).prod() ** ((60*6.5*252)/df.shape[0]) - 1
    sharpe = df['return'].mean()/df['return'].std()
    anualised_sharpe = sharpe*((60*6.5*252)**0.5)
    n_trade, revised_action = get_n_trade(df.action)
    df['real_action'] = revised_action
    signals = df[df['action']>0].shape[0]
    return df, (annualised_return, anualised_sharpe, n_trade, signals)

def get_n_trade(action):
    # 0 hold, 1 buy, 2 sell
    pos = False
    n_trade = 0
    revised_action = []
    for a in action:
        if a == 2 and pos:
            n_trade += 1
            pos = False
            revised_action.append(a)
        elif a == 1 and not pos:
            n_trade += 1
            pos = True
            revised_action.append(a)
        else:
            revised_action.append(0)
    return n_trade+1, revised_action
```

```python
def load_benchmark(ticker):
    benchmark = pd.read_csv(f'result/benchmark_{ticker}.csv')
    benchmark['cum_return'] = ((benchmark['reward']/100+1).cumprod()-1)*100
    benchmark['return'] = benchmark['reward']/100
    ar = (benchmark['reward']/100).add(1).prod() ** ((60*6.5*252)/benchmark.shape[0]) - 1
    sharpe = benchmark['return'].mean()/benchmark['return'].std()
    ash = sharpe*((60*6.5*252)**0.5)
    return benchmark, (ar, ash, 2)
```

```python
def plot_pnl(df, start_i, end_i):
    
    y_range = df['close'].iloc[start_i-10:end_i+10].max() - df['close'].iloc[start_i-10:end_i+10].min()
    arrow_head_scale = y_range * 0.005
    arrow_tail_scale = arrow_head_scale * 10
    pad = int(df['step'].iloc[start_i:end_i].shape[0] * 0.05)
    
    _, ax = plt.subplots(1,1,figsize=(20,10))
    ax.plot(df['step'].iloc[start_i-pad:end_i+pad], df['close'].iloc[start_i-pad:end_i+pad], alpha=0.7)
    
    # buy arrow
    arrow = mpatches.FancyArrowPatch(
        (df['step'].iloc[start_i], df['close'].iloc[start_i]-arrow_tail_scale),
        (df['step'].iloc[start_i], df['close'].iloc[start_i]-arrow_head_scale),
        mutation_scale=20,color='g')
    ax.add_patch(arrow)
    # sell arrow
    arrow = mpatches.FancyArrowPatch(
        (df['step'].iloc[end_i], df['close'].iloc[end_i]+arrow_tail_scale),
        (df['step'].iloc[end_i], df['close'].iloc[end_i]+arrow_head_scale),
        mutation_scale=20,color='r')
    ax.add_patch(arrow)
    
    # labels
    ax.locator_params(axis='x', nbins=6)
    ax.set_ylabel('Stock Price (USD)')
    ax.set_xlabel('Time Step (Minute)')
    
    ax2 = ax.twiny()
    ax2.set_xticks(ax.get_xticks())
    ax2.set_xbound(ax.get_xbound())
    ax2.set_xticklabels([df['time'].iloc[int(step-1)] for step in ax.get_xticks()], ha="left")
    ax2.set_title("Time")

    
```

```python
def plot_pnl_all(df, trades, max_count=10):
    
    trade_count = min(trades.shape[0], max_count)
    
    _, ax = plt.subplots(trade_count,1,figsize=(20,10*trade_count))
    
    for idx, (start_i, end_i) in enumerate(trades[:max_count]):
    
        y_range = df['close'].iloc[start_i-10:end_i+10].max() - df['close'].iloc[start_i-10:end_i+10].min()
        arrow_head_scale = y_range * 0.005
        arrow_tail_scale = arrow_head_scale * 10
        pad = int(df['step'].iloc[start_i:end_i].shape[0] * 0.05)

        ax[idx].plot(df['step'].iloc[start_i-pad:end_i+pad], df['close'].iloc[start_i-pad:end_i+pad], alpha=0.7)

        # buy arrow
        arrow = mpatches.FancyArrowPatch(
            (df['step'].iloc[start_i], df['close'].iloc[start_i]-arrow_tail_scale),
            (df['step'].iloc[start_i], df['close'].iloc[start_i]-arrow_head_scale),
            mutation_scale=20,color='g')
        ax[idx].add_patch(arrow)
        # sell arrow
        arrow = mpatches.FancyArrowPatch(
            (df['step'].iloc[end_i], df['close'].iloc[end_i]+arrow_tail_scale),
            (df['step'].iloc[end_i], df['close'].iloc[end_i]+arrow_head_scale),
            mutation_scale=20,color='r')
        ax[idx].add_patch(arrow)
```
