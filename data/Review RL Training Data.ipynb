```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
files_dir = 'fc_rl_first'
os.listdir(files_dir)
```

```python
def quick_plot(file_loc, ylabel='', xlabel='Episodes', modelname = 'DDQN_FC'):
    df = pd.read_csv(file_loc)
    df.drop('Wall time', axis=1, inplace=True)
    df.set_index(df.columns.values[0], inplace=True)
    df.plot(figsize=(8,5), ylabel=ylabel, xlabel=xlabel, legend=False)
    plt.title(f'{modelname} : {xlabel} vs {ylabel}')
    file_name = f'{modelname}_{xlabel}_{ylabel}.png'
    plt.savefig(os.path.join(files_dir, file_name))
    plt.show()
```

```python
files_paths = glob.glob(files_dir + '/*.csv')
files_paths
```
